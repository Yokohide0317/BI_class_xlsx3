```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import japanize_matplotlib

%matplotlib inline
```

### データ読み込み
※ 疑似データです。

```python
ice_df = pd.read_csv("/content/アイスと水難事故.csv")
ice_df.head()
```

### アイス vs 水難事故発生件数
- アイスの売上が上がると、水難事故が発生しやすい？

```python
sns.scatterplot(data=ice_df, x="アイス", y="水難事故")
plt.title("アイスの売上数と水難事故の件数")
plt.show()
```

### 本当は、

`アイス -> 水難事故` ではなく、<br>
`アイス <- 気温 -> 水難事故`なはず。

```python
sns.barplot(data=ice_df, x="月", y="アイス", color="gray")
plt.title("月ごとのアイスの売上数")
plt.show()
```

```python
sns.barplot(data=ice_df, x="月", y="水難事故", color="gray")
plt.title("月ごとの水難事故の件数")
plt.show()
```
