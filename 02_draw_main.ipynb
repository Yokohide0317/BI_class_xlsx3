# 円グラフを描いてみる(那覇)

```python
# 使用するデータ(info)の再確認。".head(5)"で、上の5件を取得。
info.head(5)
```

```python
# 行名を指定し、那覇を抽出 -> "naha"に代入
# ".loc[行名]"で、行を抽出。
naha = info.loc[0]
naha
```

```python
# 那覇の人口についての列名。
population_colname = ["15歳未満-人口", "15～64歳-人口", "65歳以上-人口"]

# 値を抽出。
# DataFrame[列名のリスト] で、指定の列を取得。
naha_popu = naha[population_colname].to_list()
naha_popu
```

```python
plt.pie(
    naha_popu, #データの入力
    labels=population_colname, #ラベルの指定
    counterclock=False, #時計周りに
    autopct="%1.2f%%" #小数第2位まで表示
)

plt.title("那覇市の年齢層別 人口割合(%)") #タイトルの追加
plt.show()
```

# 散布図

### 市町村ごとのFree-WiFiの数と、人口データを組み合わせる

```python
wifi.head(5)
```

### WiFiデータは、カウントデータになっていない！

```python
# -> "市区町村名"で値ごとにカウントすればいい？
## .value_counts()で、要素のカウント
wifi_counts = wifi[["市区町村名"]].value_counts()

## .value_counts()では、辞書形式が返される。
## 操作しやすいように、DataFrame形式で読み込む。
wifi_counts = pd.DataFrame(wifi_counts, columns=["wifi-count"])
wifi_counts.head()
```

```python
info.head()
```

```python
# WiFiデータとinfoデータを結合。
## 結合には、wifiデータのindex、infoデータの"市区町村名"列を指標にする。
## howは、"inner": 内部結合、"outer": 外部結合
wifi_info_df = wifi_counts.merge(info, left_index=True, right_on="市区町村名", how="inner")
wifi_info_df.head()
```

```python
# 散布図
sns.scatterplot(data=wifi_info_df, x="第３次産業事業所数", y="wifi-count")
plt.show()
```

## 那覇市の値が大きすぎて、よくわからない...

```python
# データフレームをいじって、外れ値処理する必要がある。
# が、今回は、グラフの範囲を指定して、那覇市を外してみる。

sns.scatterplot(data=wifi_info_df, x="第３次産業事業所数", y="wifi-count")
# x軸の範囲を指定 (最小, 最大)
plt.xlim((-10, 6000))
# y軸の範囲を指定 (最小, 最大)
plt.ylim((-10, 400))
plt.show()
```
